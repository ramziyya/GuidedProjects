```python
import numpy as np
import pandas as pd
import seaborn as sns


import matplotlib
import matplotlib.pyplot as plt
plt.style.use ('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\ibrra\Downloads\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df.dropna().head()
```

```python
df = df.sort_values(by = 'gross', inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.drop_duplicates().head()
```

```python
df.head()
```

```python
plt.scatter (x = df['budget'], y = df['gross'] ) 

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws= {'color':'red'}, line_kws={'color':'blue'})
```

```python
df[['year', 'score', 'budget', 'gross', 'runtime', 'runtime']].corr()
```

```python
correlation_matrix = df[['year', 'score', 'budget', 'gross', 'runtime', 'runtime']].corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Values')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for column in df_numerized.columns:
    if(df_numerized[column].dtype == 'object'):
        df_numerized[column] = df_numerized[column].astype('category') 
        df_numerized[column] = df_numerized[column].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized[['year', 'score', 'budget', 'gross', 'runtime', 'runtime']].corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)


plt.title('Correlation Matrix for Numeric Values')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized[['year', 'score', 'budget', 'gross', 'runtime', 'runtime']].corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)


plt.title('Correlation Matrix for Numeric Values')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
